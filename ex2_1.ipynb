```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

figsize = (10, 10)

```

```python

img = np.zeros((50, 50))
img[20:30, 20:30] = 1

plt.figure(figsize=figsize)
plt.imshow(img,cmap="gray")
plt.show()

```

```python
kernel = np.zeros((5,5),dtype=np.uint8)
kernel[2,:] = 1
kernel[:,2] = 1


plt.figure(figsize=figsize)
plt.imshow(kernel,cmap="gray")
plt.show()

```

```python
def my_dilate(img,kernel):
    #TODO: build dilate function without cv2.dilate
    img_new = cv2.filter2D(img, ddepth=-1, kernel=kernel)     
    T, img_new = cv2.threshold(img_new,0,1, cv2.THRESH_BINARY)  
    return img_new

plt.figure(figsize=figsize)
plt.imshow(my_dilate(img,kernel),cmap="gray")
plt.show()

```

```python
dilated = cv2.dilate(img,kernel)
my_dilated = my_dilate(img,kernel)
#plt.figure(figsize=figsize)
#plt.imshow(dilated,cmap="gray")
print(dilated[19])
print(my_dilated[19])
print(dilated.shape)
print(my_dilated.shape)
```

```python
# TODO: show that cv2.dilate and my_dilate are the same using absolute difference
comparison = cv2.dilate(img,kernel) == my_dilate(img, kernel)
if comparison.all():
    print("cv2.dilate & my_dilate are the same!")
else: 
    print("try again...")

```

```python
def my_erode(img,kernel):
    #TODO: build erode function without cv2.erode
    img_new = cv2.filter2D(img, ddepth=-1, kernel=kernel)  
    _,img_new = cv2.threshold(img_new, 8 ,1, cv2.THRESH_BINARY)
    return img_new 

plt.figure(figsize=figsize)
plt.imshow(my_erode(img,kernel),cmap="gray")
plt.show()

```

```python
eroded_cv2 = cv2.erode(img,kernel)
plt.figure(figsize=figsize)
plt.imshow(eroded_cv2,cmap="gray")
plt.show()

```

```python
# TODO: show that cv2.erode and my_erode are the same using absolute difference
comparison = cv2.erode(img,kernel) == my_erode(img, kernel)

if comparison.all():
    print("cv2.erode & my_erode are the same!")
else: 
    print("try again...")

```
